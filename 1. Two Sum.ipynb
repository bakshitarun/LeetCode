```
Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.
You may assume that each input would have exactly one solution, and you may not use the same element twice.
You can return the answer in any order.

Example 1:
Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].
Example 2:

Input: nums = [3,2,4], target = 6
Output: [1,2]


Constraints:
2 <= nums.length <= 104
-109 <= nums[i] <= 109
-109 <= target <= 109
Only one valid answer exists.
 

Follow-up: Can you come up with an algorithm that is less than O(n2) time complexity?
```

# Approach 1

Creating a nested for loop. In the first loop we are looping over the length of the list and in the next one we are iterating over the length of list but we start from i+1. Next add the i,j positions in the list to check if it matches the target and if it does then we return the list

Time Complexity: O(n2)

````python
from typing import List
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for i in range(len(nums)):
                for j in range(i+1,len(nums)):
                    if nums[i]+ nums[j] == target:
                        return [i,j]

````

````python
b = Solution()
b.twoSum([1,2,3,4],7)
````

#Approach 2
We are using a hashmap to reduce the time complexity to O(n2). We do this by going through the list of values one time. Every time we iterate we subtract the current value from target value and add it to the hashmap. In every iteration we check whether we have a value in the hashmap or not

For Example:
our_list = [1,2,3,4,5]
target = 9

Iteration1:
hashmap = {}
return if 8 is present in hashmap else add value to hashmap
complement = 9-1 = 8

Iteration2:
hashmap = {1: 0}
return if 7 is present in hashmap else add value to hashmap
complement = 9-2 = 7

Iteration3:
hashmap = {1: 0, 2: 1}
return if 6 is present in hashmap else add value to hashmap
complement = 9-3 = 6

Iteration4:
hashmap = {1: 0, 2: 1, 3: 2}
return if 5 is present in hashmap else add value to hashmap
complement = 9-4 = 5

Iteration5:
hashmap = {1: 0, 2: 1, 3: 2, 4: 3}
return if 4 is present in hashmap else add value to hashmap






````python
from typing import List
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        hashmap = {}
        for i in range(len(nums)):
            complement = target - nums[i]
            if complement in hashmap:
                return [i, hashmap[complement]]
            hashmap[nums[i]] = i
  
````

````python
result = Solution()
result.twoSum([1,2,3,4],7)
````
